### Correlation between Lenghts
How correlation works?
we use the corr() function to see how 2 attributes are related to each other
output will be a nxn matrix with values[-1,1]

+1 → perfect positive relation (they increase together).

0 → no linear relation.

–1 → perfect negative relation.

```python
fish[['Length1','Length2','Length3']].corr()
```

These correlation values (~0.99) show near-perfect positive correlation among Length1, Length2, and Length3.

Implication:

They carry identical information.

Keeping all three adds redundancy and causes multicollinearity, which can distort model weights (especially in linear or logistic regression).

```python
# Now we can take average of the lenghts and add the columns
average_length = fish[['Length1','Length2','Length3']].mean(axis=1)
fish['Average_Length'] = average_length
print(fish.groupby('Species')['Average_Length'].mean().sort_values(ascending=False))
```
